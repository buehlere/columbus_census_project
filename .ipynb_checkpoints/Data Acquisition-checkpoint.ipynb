```python
years = [2009,2010,2011,2012,2013,2014,2015,2016,2017]

api_key = "6518d805990d22f00cb5e9573e86376f70425c04"
url = "https://api.census.gov/data/2017/acs/acs1?get=NAME,B25034_010E&for=county:049&in=state:39&key=" + api_key  
url = "https://api.census.gov/data/2017/acs/acs5?get=B00001_001E,B25056_001E,NAME&for=tract:*&in=state:39%20county:049&key=" + api_key
response = requests.get(url)

#06000US3904918000
query = json.loads(response.text)
print(response)
pp.pprint(query)
```

```python
df = pd.DataFrame(query)
new_header = df.iloc[0] #grab the first row for the header
df = df[1:] #take the data less the header row
df.columns = new_header
df.head()
```

```python
df['NAMELSAD'], df['County_State'] = df['NAME'].str.split(',', 1).str
df = df.drop(['NAME'],axis=1)
df['County'], df['State'] = df['County_State'].str.split(',', 1).str
df = df.drop(['County_State'],axis=1)
df.head()
```

```python
ohio = gpd.read_file('tl_2018_39_tract/tl_2018_39_tract.shp')
ohio = ohio[ohio['COUNTYFP']== '049']
ohio.head()
```

```python
combind = pd.concat([df.set_index('NAMELSAD'),ohio.set_index('NAMELSAD')], axis=1, join='inner')
combind.reset_index(level=0, inplace=True)
combind.head()
```

```python
combind = gpd.GeoDataFrame(combind, geometry='geometry')
combind.to_file('MyGeometries.shp', driver='ESRI Shapefile')
```
